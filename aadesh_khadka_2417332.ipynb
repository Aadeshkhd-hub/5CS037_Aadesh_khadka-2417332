1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.

```python
Q1_score = filtered_df['score'].quantile(0.25)
Q3_score = filtered_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

outliers_score = filtered_df[(filtered_df['score'] < lower_bound_score) | (filtered_df['score'] > upper_bound_score)]

#  Detecting outliers for 'Log GDP per Capita' using 1.5 * IQR rule
Q1_gdp = filtered_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = filtered_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

outliers_gdp = filtered_df[(filtered_df['Log GDP per capita'] < lower_bound_gdp) | (filtered_df['Log GDP per capita'] > upper_bound_gdp)]

outliers = pd.concat([outliers_score, outliers_gdp]).drop_duplicates()
(outliers[['Country name', 'score', 'Log GDP per capita']])
```

Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.

```python

outliers_score = filtered_df[(filtered_df['score'] < lower_bound_score) | (filtered_df['score'] > upper_bound_score)]
outliers_gdp = filtered_df[(filtered_df['Log GDP per capita'] < lower_bound_gdp) | (filtered_df['Log GDP per capita'] > upper_bound_gdp)]

outliers = pd.concat([outliers_score, outliers_gdp]).drop_duplicates()

plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['Log GDP per capita'], filtered_df['score'], label='Normal df', color='blue', alpha=0.6)

plt.scatter(outliers['Log GDP per capita'], outliers['score'], label='Outliers', color='red', s=100, edgecolor='black')

plt.xlabel('Log GDP per Capita')
plt.ylabel('score')
plt.title('Scatter Plot: Log GDP per Capita vs Score (Outliers Highlighted)')

plt.legend()

plt.show()
```

```python
south_asia_range = filtered_df['score'].max() - filtered_df['score'].min()
south_asia_mean = filtered_df['score'].mean()
south_asia_std = filtered_df['score'].std()
south_asia_cv = (south_asia_std / south_asia_mean) * 100

# Compute Range and CV for Middle East
middle_east_range = middle_eastern_df['score'].max() - middle_eastern_df['score'].min()
middle_east_mean = middle_eastern_df['score'].mean()
middle_east_std = middle_eastern_df['score'].std()
middle_east_cv = (middle_east_std / middle_east_mean) * 100

# Output the results
print(f"South Asia - Range: {south_asia_range}, CV: {south_asia_cv:.2f}%")
print(f"Middle East - Range: {middle_east_range}, CV: {middle_east_cv:.2f}%")
if south_asia_cv>middle_east_cv:
    print("South asia has greater variability in happiness")
else:
    print("Middle eastern has greater variability in happiness")
```
